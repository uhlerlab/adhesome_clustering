### Construct intraX edge list

```python
# Create edge list containing intraX edges between active loci
chr_list = np.arange(1,22+1,1)
edge_list_intra0 = pd.DataFrame(columns=['locus1','locus2','hic_contact','chr'])
for chrom in tqdm(chr_list):
    time.sleep(.01)
    subindex = ['chr_'+str(chrom)+'_'+'loc_'+str(loc) for loc in active_loci_dict[chrom]]
    # Load HiC data for this chromosome pair
    processed_hic_filename = 'hic_'+'chr'+str(chrom)+'_'+'chr'+str(chrom)+'_norm1_filter3'+'.pkl'
    hic_chpair_df = pickle.load(open(dir_processed_hic+processed_hic_filename, 'rb'))
    hic_chpair_active_df = hic_chpair_df.loc[active_loci_dict[chrom],active_loci_dict[chrom]]
    hic_chpair_active_df.index = subindex
    hic_chpair_active_df.columns = subindex
    # Hi-C data in key-value format
    new_index = pd.MultiIndex.from_tuples(itertools.combinations(subindex,2), names=["locus1","locus2"])
    hic_chpair_df1 = hic_chpair_active_df.stack().reindex(new_index).reset_index(name='hic_contact')
    hic_chpair_df1['chr'] = chrom
    hic_chpair_df1 = hic_chpair_df1[hic_chpair_df1['hic_contact']>0]
    # Create network from edge list
    edge_list_intra0 = pd.concat([edge_list_intra0,hic_chpair_df1])
```

```python
# Select top 0.1% of intraX edges (0.1% for each chromosome)
edge_list_intra = pd.DataFrame(columns=['locus1','locus2','hic_contact','significant'])
for chrom in tqdm(np.arange(1,22+1,1)):
    time.sleep(.01)
    edge_list_chrom = edge_list_intra0[edge_list_intra0['chr']==chrom][['locus1','locus2','hic_contact']]
    q = np.quantile(edge_list_chrom['hic_contact'].values,0.999)
    edge_list_chrom['significant'] = (edge_list_chrom['hic_contact']>=q)
    # Update edge_list_raw_intra
    edge_list_intra = pd.concat([edge_list_intra,edge_list_chrom])
```

```python
# Plot histogram of intraX edge weights (Hi-C contact values) between active loci
print('Total number of intraX edges = '+str(len(edge_list_intra)))
print('Total number of significant intraX edges = '+str(len(edge_list_intra[edge_list_intra['significant']==True]))+' ('+str(100*np.round(len(edge_list_intra[edge_list_intra['significant']==True])/len(edge_list_intra),3))+'%)')
plt.figure()
plt.hist(edge_list_intra['hic_contact'], bins = np.arange(0,10,0.2), 
         color='blue', alpha=0.5, label='intraX edges')
plt.hist(edge_list_intra[edge_list_intra['significant']==True]['hic_contact'], bins = np.arange(0,10,0.2), 
         color='green', alpha=0.5, label='significant intraX edges')
#plt.vlines(3, ymin=0, ymax=100000, color='red')
plt.xlabel('hic contact (edge weight)')
plt.ylabel('intraX edge count')
plt.yscale('log')
plt.legend()
plt.title('Histogram of intraX edge weights')
plt.show()
```

```python
resolution_ls = np.arange(0.01,0.26,0.02)
modularities = []
num_comm = []
average_nodes_comm = []
num_adh_comm = []
average_nodes_adh_comm = []
average_adh_nodes_adh_comm = []
for r in tqdm(resolution_ls):
    time.sleep(.01)
    # Cluster network using Louvain community detection
    partition = community.best_partition(G, resolution=r, weight='loghic', random_state=6)
    partition_str = {key: str(val) for i,(key,val) in enumerate(partition.items())}
    nx.set_node_attributes(G, partition, 'modularity')
    nx.set_node_attributes(G, partition_str, 'modularity_str')
    n_communities = len(set(partition.values()))
    # Summary of each cluster
    community_summary = pd.DataFrame(columns=['community', 'n_nodes', 'chroms','n_edges','n_genes','n_tf','tf_list','n_adhesome','adhesome_list'])
    for comm in range(n_communities):
        sub_nodes = [locus for locus in partition.keys() if partition[locus]==comm]
        sub_attributes = node_attributes[node_attributes.index.isin(sub_nodes)]
        subG = G.subgraph(sub_nodes)
        n_nodes = len(subG.nodes)
        chroms = np.unique(sub_attributes['chr'].sum())
        n_edges = len(subG.edges)
        n_genes = sub_attributes['has_genes'].sum()
        n_tf = len(np.unique(sub_attributes['locus_tf_gene_list'].sum()))
        tf_list = np.unique(sub_attributes['locus_tf_gene_list'].sum())
        n_adhesome = len(np.unique(sub_attributes['locus_adhesome_gene_list'].sum()))
        adhesome_list = np.unique(sub_attributes['locus_adhesome_gene_list'].sum())
        community_summary.loc[comm] = [comm, n_nodes,chroms,n_edges,n_genes,n_tf,tf_list,n_adhesome,adhesome_list]
    community_summary = community_summary.sort_values(by='n_adhesome', ascending=False)
    modularities.append(community.modularity(partition,G))
    num_comm.append(len(set(partition.values())))
    average_nodes_comm.append(community_summary['n_nodes'].mean())
    num_adh_comm.append(len(community_summary[community_summary['n_adhesome']>0]))
    average_nodes_adh_comm.append(community_summary[community_summary['n_adhesome']>0]['n_nodes'].mean())
    average_adh_nodes_adh_comm.append(community_summary[community_summary['n_adhesome']>0]['n_adhesome'].mean())
```

### Random walk
We use the technique developed in:
- https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4157572/pdf/nihms-623050.pdf
- https://pubmed.ncbi.nlm.nih.gov/24482476/

- <span class="mark">It is better to adopt a first passage method (ask Shiva for the paper)</span>

```python
# We consider the giant component
G = G_saved.copy()
print('Total number of nodes in G = '+str(len(G.nodes)))
print('Number of active adhesome nodes in G = '+str(len(set(active_adhesome_loci).intersection(set(G.nodes)))))
print('Total number of edges in G = '+str(len(G.edges)))
print('Average node degree in G = '+str(np.mean(list(dict(G.degree).values()))))
print('Diameter of G = '+str(nx.diameter(G)))
```

```python
# Parameters of the random walk
d_list = np.arange(10,40,5)
n_repeat = 100
adhesome_loci_in_network = [locus for locus in G.nodes if (locus in active_adhesome_loci)]
all_loci_in_network = G.nodes
```

```python
random.seed(13)
# Loop over different depths
depth_loop_summary_df = pd.DataFrame(0, index=d_list, columns=all_loci_in_network)
for d in tqdm(d_list):
    time.sleep(.01)
    # Repeat RW procedure n_repeat times
    repeat_loop_summary_df = pd.DataFrame(0, index=np.arange(0,n_repeat), columns=all_loci_in_network)
    for i in np.arange(n_repeat):
        # Define an empty dataframe to count visits
        visit_df = pd.DataFrame(0, index=adhesome_loci_in_network, columns=all_loci_in_network)
        # Loop over all adhesome nodes
        for seed in adhesome_loci_in_network:
            u = seed
            visit_df.loc[seed,u] += 1
            # Perform RW and update the visit time of each node in visit_df
            for j in np.arange(d):
                u = random.choice(list(G.neighbors(u)))
                visit_df.loc[seed,u] += 1
        repeat_loop_summary_df.loc[i,:] = visit_df.sum(axis=0)/visit_df.sum().sum()
    depth_loop_summary_df.loc[d,:] = repeat_loop_summary_df.mean(axis=0)
pickle.dump(depth_loop_summary_df, open(saving_dir+'depth_loop_summary_df2.pkl', 'wb'))
```

```python
# Plot distance between visit frequencies for consecutive depths
#depth_loop_summary_df = pickle.load(open(saving_dir+'depth_loop_summary_df.pkl', 'rb'))
differences_between_rows = [np.linalg.norm(depth_loop_summary_df.iloc[i,:]-depth_loop_summary_df.iloc[i-1,:],1)
                            for i in range(1,len(depth_loop_summary_df))]
# Plot
plt.figure()
plt.plot(range(1,len(depth_loop_summary_df)), differences_between_rows)
plt.xlabel('iteration')
plt.ylabel('L1 difference of node frequencies')
plt.yscale('log')
plt.show()
```

```python
# Now that we have determined the optimal depth, run RW and count the frequency of edges traversed
list_edges = list(G.edges)
nx.set_edge_attributes(G, {list_edges[i]: 0 for i in range(len(G.edges))},'count')
nx.set_edge_attributes(G, {list_edges[i]: 0 for i in range(len(G.edges))},'norm_count')
```

```python
# Run procedure
random.seed(13)
d = 40
# Repeat RW procedure n_repeat times
summary_df = pd.DataFrame(0, index=np.arange(0,n_repeat), columns=all_loci_in_network)
for i in tqdm(np.arange(n_repeat)):
    time.sleep(.01)
    # Define an empty dataframe to count visits
    node_visit_df = pd.DataFrame(0, index=adhesome_loci_in_network, columns=all_loci_in_network)
    # Loop over all adhesome nodes
    for seed in adhesome_loci_in_network:
        u = seed
        node_visit_df.loc[seed,u] += 1
        # Perform RW and update the visit time of each node in visit_df
        for j in np.arange(d):
            v = random.choice(list(G.neighbors(u)))
            node_visit_df.loc[seed,v] += 1
            G[u][v]['count'] += 1
            u = v
    summary_df.loc[i,:] = node_visit_df.sum(axis=0)/node_visit_df.sum().sum()
```

```python
# Summarize procedure

# Add visit frequency as node attribute
final_node_summary_df = summary_df.mean(axis=0)
final_node_summary_index = list(final_node_summary_df.index)
nx.set_node_attributes(G, {final_node_summary_index[i]: final_node_summary_df[final_node_summary_index[i]] 
                           for i in range(len(final_node_summary_index))}, 
                       'visit_freq')

# Add normalized edge visit count as edge attribute
edge_count_attribute = list(nx.get_edge_attributes(G, 'count').values())
max_count = max(edge_count_attribute)
for edge in G.edges:
    G[edge[0]][edge[1]]['norm_count'] = G[edge[0]][edge[1]]['count']/max_count
```

- https://proceedings.neurips.cc/paper/2003/file/87682805257e619d49b8e0dfdc14affa-Paper.pdf

See also:
- https://science.sciencemag.org/content/sci/suppl/2014/01/30/343.6170.506.DC1/Novarino.SM.pdf
- https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1000641

```python
# Form affinity matrix of G
W = nx.adjacency_matrix(G, weight=None).todense()
```

```python
# Construct normalized matrix S
degrees = np.array(W.sum(axis=1).flat)
inv_sq_D = np.diag(np.power(degrees,-0.5))
S = np.linalg.multi_dot([inv_sq_D,W,inv_sq_D])
```

```python
# Response
Y = np.matrix([int(locus in adhesome_loci_in_network) for locus in G.nodes]).reshape(len(G.nodes),1)
Y.index = list(G.nodes)
```
